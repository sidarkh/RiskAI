# RiskAI – Modelltraining und -vergleich

In dieser Phase trainieren wir verschiedene Machine-Learning-Modelle zur Erkennung von Kreditkartenbetrug.
Wir vergleichen:
1. Logistische Regression
2. Entscheidungsbäume
3. Random Forest (später)
4. Gradient Boosting / XGBoost (später)


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc

X_train = pd.read_csv("../data/processed/X_train_scaled.csv")
X_test = pd.read_csv("../data/processed/X_test_scaled.csv")
y_train = pd.read_csv("../data/processed/y_train.csv").values.ravel()
y_test = pd.read_csv("../data/processed/y_test.csv").values.ravel()

print(f"Trainingsdaten: {X_train.shape} | Testdaten: {X_test.shape}")
```

```python
# 1. Baseline Logistische Regression (Standard, ohne Gewichtung)
lr_baseline = LogisticRegression(max_iter=1000, random_state=42)
lr_baseline.fit(X_train, y_train)

# 2. Logistische Regression mit ausbalancierten Klassengewichten
lr_balanced = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
lr_balanced.fit(X_train, y_train)


print("Beide logistischen Regressionsmodelle erfolgreich trainiert!")
```

```python
# Wir vergleichen und bewerten jetzt unsere Modelle
y_pred_baseline = lr_baseline.predict(X_test)
y_pred_balanced = lr_balanced.predict(X_test)


print("=== BASELINE MODELL (Ohne Klassengewichte) ===")
print(classification_report(y_test, y_pred_baseline))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_baseline))


print("=== BASELINE MODELL (Mit class_weight=balanced) ===")
print(classification_report(y_test, y_pred_balanced))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_balanced))
```

## Entscheidungsbäume (Decision Trees)

Wir trainieren nun Entscheidungsbäume (Baseline und Balanced).
Um Overfitting zu verhindern, begrenzen wir die maximale Tiefe des Baumes auf `max_depth=5`.


```python
dt_baseline = DecisionTreeClassifier(max_depth=5, random_state=42)
dt_baseline.fit(X_train, y_train)

dt_balanced = DecisionTreeClassifier(max_depth=5, class_weight="balanced", random_state=42)
dt_balanced.fit(X_train, y_train)

print("Entscheidungsbäume erfolgreich trainiert.")
```

```python
y_pred_dt_baseline = dt_baseline.predict(X_test)
y_pred_dt_balanced= dt_balanced.predict(X_test)

print("=== DECISION TREE BASELINE (Ohne Klassengewichte) ===")
print(classification_report(y_test, y_pred_dt_baseline))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_dt_baseline))

print("\n=== DECISION TREE BALANCED (Mit class_weight='balanced') ===")
print(classification_report(y_test, y_pred_dt_balanced))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_dt_balanced))
```

## Random Forest Classifier

Wir trainieren nun einen Random Forest mit 100 Bäumen.
Um das Training auf deinem Mac zu beschleunigen, nutzen wir `n_jobs=-1` (nutzt alle verfügbaren CPU-Kerne parallel).


```python
from sklearn.ensemble import RandomForestClassifier


rf_baseline = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_baseline.fit(X_train, y_train)

rf_balanced = RandomForestClassifier(n_estimators=100, max_depth=10, class_weight='balanced', random_state=42, n_jobs=-1)
rf_balanced.fit(X_train, y_train)

print("Random Forests erfolgreich trainiert!")
```

```python
y_pred_rf_baseline = rf_baseline.predict(X_test)
y_pred_rf_balanced = rf_balanced.predict(X_test)

print("=== RANDOM FOREST BASELINE (Ohne Klassengewichte) ===")
print(classification_report(y_test, y_pred_rf_baseline))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_rf_baseline))

print("\n=== RANDOM FOREST BALANCED (Mit class_weight='balanced') ===")
print(classification_report(y_test, y_pred_rf_balanced))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_rf_balanced))
```

## XGBoost Classifier (Gradient Boosting)

Wir implementieren nun XGBoost. 
Für die Gewichtung berechnen wir das genaue Verhältnis von legitimen Transaktionen zu Betrugsfällen (`scale_pos_weight`).


```python
from xgboost import XGBClassifier

xgb_baseline = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, n_jobs=-1)
xgb_baseline.fit(X_train, y_train)

ratio = len(y_train[y_train == 0] / len(y_train[y_train == 1]))

xgb_balanced = XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, scale_pos_weight=ratio, random_state=42, n_jobs=-1)
xgb_balanced.fit(X_train, y_train)

print(f"XGBoost Modelle erfolgreich trainiert! (Gewichtungsfaktor scale_pos_weight: {ratio:.2f})")
```

```python

y_pred_xgb_baseline = xgb_baseline.predict(X_test)
y_pred_xgb_balanced = xgb_balanced.predict(X_test)

print("=== XGBOOST BASELINE (Ohne Klassengewichte) ===")
print(classification_report(y_test, y_pred_xgb_baseline))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_xgb_baseline))

print("\n=== XGBOOST BALANCED (Mit scale_pos_weight) ===")
print(classification_report(y_test, y_pred_xgb_balanced))
print("Konfusionsmatrix:")
print(confusion_matrix(y_test, y_pred_xgb_balanced))
```

## Modellbewertung: ROC-AUC vs. PR-AUC

Da unser Datensatz extrem unbalanciert ist, vergleichen wir die Modelle anhand der ROC- und der Precision-Recall-Kurve. 
Besonders die Precision-Recall-Kurve (PR-AUC) zeigt uns die verlässliche Leistung des Modells bei der Minderheitsklasse.


```python
from sklearn.metrics import roc_curve, precision_recall_curve, auc

# Wahrscheinlichkeiten für die positive Klasse (Klasse 1) vorhersagen
y_prob_lr = lr_balanced.predict_proba(X_test)[:, 1]
y_prob_rf = rf_balanced.predict_proba(X_test)[:, 1]
y_prob_xgb = xgb_balanced.predict_proba(X_test)[:, 1]

# ROC-Kurven und ROC-AUC berechnen
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)

roc_auc_lr = auc(fpr_lr, tpr_lr)
roc_auc_rf = auc(fpr_rf, tpr_rf)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)

# Precision-Recall-Kurven und PR-AUC berechnen
prec_lr, rec_lr, _ = precision_recall_curve(y_test, y_prob_lr)
prec_rf, rec_rf, _ = precision_recall_curve(y_test, y_prob_rf)
prec_xgb, rec_xgb, _ = precision_recall_curve(y_test, y_prob_xgb)

pr_auc_lr = auc(rec_lr, prec_lr)
pr_auc_rf = auc(rec_rf, prec_rf)
pr_auc_xgb = auc(rec_xgb, prec_xgb)

# Diagramme zeichnen
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 1. ROC-Kurven plotten
ax1.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.3f})')
ax1.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.3f})')
ax1.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {roc_auc_xgb:.3f})')
ax1.plot([0, 1], [0, 1], 'k--', label='Zufall (AUC = 0.500)')
ax1.set_title('Vergleich der ROC-Kurven')
ax1.set_xlabel('False Positive Rate (Falsch-Positiv-Rate)')
ax1.set_ylabel('True Positive Rate (Recall / Trefferquote)')
ax1.legend(loc='lower right')

# 2. Precision-Recall-Kurven plotten
ax2.plot(rec_lr, prec_lr, label=f'Logistic Regression (AUC = {pr_auc_lr:.3f})')
ax2.plot(rec_rf, prec_rf, label=f'Random Forest (AUC = {pr_auc_rf:.3f})')
ax2.plot(rec_xgb, prec_xgb, label=f'XGBoost (AUC = {pr_auc_xgb:.3f})')
ax2.set_title('Vergleich der Precision-Recall-Kurven (Beste Metrik)')
ax2.set_xlabel('Recall (Trefferquote)')
ax2.set_ylabel('Precision (Verlässlichkeit)')
ax2.legend(loc='lower left')

plt.tight_layout()
plt.show()

print(f"PR-AUC - Logistic Regression: {pr_auc_lr:.4f}")
print(f"PR-AUC - Random Forest:      {pr_auc_rf:.4f}")
print(f"PR-AUC - XGBoost:            {pr_auc_xgb:.4f}")
```
